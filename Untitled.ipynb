```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
sns.set_style("darkgrid")
```

```python
#dataset 읽어들이기
train=pd.read_csv("train.csv")
train.head()
#train.isnull().sum()
pd.set_option('max_row',1000)
```

```python
train
#target variable !!!!!!!!!!!!!!! count = casual + registered( 자전거 총 대여량)

#

casual= 가입 안하고 빌린 사람들
#registered=가입되어 있는 빌린 사람들
#workingday = 공휴일도 아니고 주말도 아닌 날
#weather = 1~4로 표시된 날씨
#temp=실제기온
#atemp=체감온도
```

```python
#timestamp를 연원일 시분초로 쪼개기
import datetime
train.datetime = train.datetime.map(lambda x: datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S'))
```

```python
train.columns
```

```python
#create new columns for date and time

# year
train['year'] = train.datetime.map(lambda x : x.year)
# month
train['month']=train.datetime.map(lambda x: x.month)
# day
train['day'] = train.datetime.map(lambda x : x.day)
# hour
train['hour'] = train.datetime.map(lambda x : x.hour)
# weekday
train['wday'] = train.datetime.map(lambda x : x.weekday())
# date
train['date'] = train.datetime.map(lambda x : x.date())
```

```python
#count의 분포 알아보기
sns.distplot(train['count'])
#0~200 사이의 값이 많음 (right skewed)
```

```python
print("Skewness: %f"%train['count'].skew())
print("kurtosis: %f"%train["count"].kurt())
```

```python
sns.boxplot(train["count"])
#outliers을 볼 수 있습니다 ==> 제거해야겠네요.
```

```python
# 각 categorical 독립변수에 따른 count의 분포를 보고자 합니다.
fig,axes = plt.subplots(ncols=2 ,nrows=2)
fig.set_size_inches(15,15)
sns.boxplot(data=train,x='season',y='count',ax=axes[0][0])
sns.boxplot(data=train,x='holiday',y='count',ax=axes[0][1])
sns.boxplot(data=train,x='workingday',y='count',ax=axes[1][0])
sns.boxplot(data=train,x='weather',y='count',ax=axes[1][1])

#의외로 봄에 자전거 사용빈도가 더 낮아요
#weather=4 : Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog  - 역시 아웃라이어 빼고는 사용을 하지 않음
#holiday일 때 더 많이 자전거가 대여되긴 하였으나 공휴일이 아닐 때와 큰 차이는 없어 보입니다.
```

```python
#month, day에 따른 count의 분포?

fig,axes = plt.subplots(ncols=3 ,nrows=2)
fig.set_size_inches(20,7)
sns.boxplot(data=train,x='month',y='count', ax=axes[0][0])
sns.boxplot(data=train,x='month',y='registered', ax=axes[0][1])
sns.boxplot(data=train,x='month',y='casual', ax=axes[0][2])
sns.boxplot(data=train,x='day',y='count', ax=axes[1][0])
sns.boxplot(data=train,x='day',y='registered', ax=axes[1][1])
sns.boxplot(data=train,x='day',y='casual', ax=axes[1][2])

#month별 분포를 보면 casual users는 겨울에는 사용빈도가 현저히 낮은 반면 registered또한 그런 경향을 보이긴 하나 큰 차이는 없습니다.
#day별 분포를 보면 registered and casual 간의 딱히 유의미한 차이가 있는 것 같지는 않습니다.
```

```python
#hour에 따른 count 분포
fig.set_size_inches(40,10)
sns.boxplot(data=train, x="hour",y="count")
```

```python
#hour에 따른 registered, casual
fig2,axes2 = plt.subplots(ncols=2 ,nrows=1)
fig2.set_size_inches(40,10)
sns.boxplot(data=train, x="hour",y="registered", ax=axes2[0])
sns.boxplot(data=train, x="hour", y="casual", ax=axes2[1])

#시간대에 따른 registered의 자전거 사용과 casual의 자전거 사용이 다른 경향을 보이네요.
#registered : 오전 7~8시, 오후 16~19시에 자전거 이용을 많이 함 : 자전거로 출퇴근 하는 것 같습니다.
#casual : 오전 9~오후 20시까지 많이 사용, 특히 시간의 경과에 따라 증가했다 감소함을 알 수 있습니다.
#count : 전체적으로 오전 출근시간대, 오후 퇴근시간대에 가장 많은 분포를 보임 - casual users < registered users 때문인 것 같아요.
```

```python
train.plot.scatter(x="temp", y="count",ylim=(0,1000))
```

```python
#continuous variable과의 관계를 보기 위해 lmplot을 그리려 하였으나 subplot으로 2*2를 그릴 수 없네요ㅠㅠㅠㅠ

fig,axes5 = plt.subplots(ncols=2,nrows=2)
fig.set_size_inches(15,10)
sns.lmplot(x="temp", y='count', data=train, fit_reg=False, ax=axes5[0][0])
sns.lmplot(x="atemp", y='count', data=train, fit_reg=False, ax=axes5[0][1])
sns.lmplot(x="humidity", y='count', data=train, fit_reg=False, ax=axes5[1][0])
sns.lmplot(x="windspeed", y='count', data=train, fit_reg=False, ax=axes5[1][1])
```

```python
#subplot을 그릴 수 있는 regplot으로 시도
#regplot : Plot data and a linear regression model fit.


fig,axes = plt.subplots(ncols=2,nrows=2)
fig.set_size_inches(15,10)
sns.regplot(data=train,x='temp',y='count', ax=axes[0][0])
sns.regplot(data=train,x='atemp',y='count', ax=axes[0][1])
sns.regplot(data=train,x='humidity',y='count',fit_reg=False, ax=axes[1][0])
sns.regplot(data=train,x='windspeed',y='count', ax=axes[1][1])
plt.show()

#그런데 windspeed와 count는 음의 상관관계를 가지고 있는데 저렇게 regression line이 양의 기울기를 갖는건
#왜 그런지 생각해봐야 할 것 같아요.

#temp and atemp look alike. ==> correlation matrix을 그려봐야겠습니다.
```

```python
ctnsvar=train[["temp","atemp","windspeed","humidity"]]
corr=ctnsvar.corr()
corr

#heatmap 
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns,
           annot=True,
           vmax=3,
           cmap="BuPu")
#temp and atemp are highly correlated : 온도와 체감온도라서 그런가 봅니다.
```
