```python
df=pd.read_csv("results3.csv")
```

```python
df.loc[df['급여'].str.contains("시급")] #4489 
```

```python
df['급여'] = df['근무지역'].apply(lambda x: x[5:9])
```

```python
sig=df[df['급여'].str.contains("시급")]
```

```python
sig['급여']=sig['급여'].apply(lambda x: x[3:])
```

```python
sig['급여']=sig['급여'].str.replace(',','')
```

```python
sig["급여"] = sig['급여'].astype(float)
sig['급여']
```

```python
sig.head()
```

```python
table = pd.pivot_table(sig, values=['급여'], index=['근무지역', '업직종'], aggfunc={
    '급여': [min, max, np.mean, len]
})
```
